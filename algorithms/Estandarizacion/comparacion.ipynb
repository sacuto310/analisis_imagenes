```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

image_data_A = nib.load('../images/T1.nii.gz').get_fdata()
image_data_B = nib.load('../images/T1_V2.nii.gz').get_fdata()
```

```python
plt.hist(image_data_A[image_data_A>10].flatten(), 100, alpha=0.5)
plt.hist(image_data_B[image_data_B>10].flatten(), 100, alpha=0.5)
plt.show
```

```python
mean_value = image_data_A.mean()
standard_deviation_value = image_data_A.std()
image_data_rescaled_A = (image_data_A - mean_value) / (standard_deviation_value)

mean_value = image_data_B.mean()
standard_deviation_value = image_data_B.std()
image_data_rescaled_B = (image_data_B - mean_value) / (standard_deviation_value)
```

```python
plt.hist(image_data_rescaled_A[image_data_rescaled_A>0.01].flatten(), 100, alpha=0.5)
plt.hist(image_data_rescaled_B[image_data_rescaled_B>0.01].flatten(), 100, alpha=0.5)
plt.show
```
